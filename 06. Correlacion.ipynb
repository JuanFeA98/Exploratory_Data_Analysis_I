## Librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Datos

```python
from sklearn.datasets import load_breast_cancer
```

```python
df = load_breast_cancer()
```

```python
data = df.data
features = df.feature_names
```

```python
df = pd.DataFrame(data, columns=features)
```

```python
df.info()
```

```python
wf = df.iloc[:, :6]
```

```python
wf
```

```python
wf.corr()
```

```python
sns.heatmap(wf.corr(), annot=True)
```

```python
corr = pd.DataFrame(wf.corr().unstack()).reset_index()
```

```python
corr.rename(columns={0: 'Correlacion'}).sort_values('Correlacion', ascending=False).reset_index(drop=True)
```

```python
sns.pairplot(wf)
plt.show()
```
