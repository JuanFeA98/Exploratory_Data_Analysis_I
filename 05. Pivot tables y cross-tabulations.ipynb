## Pivot tables y cross-tabulations

### Librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Datos

```python
df = pd.read_csv('./Data/chicago.csv')
```

```python
df['CRASH_DATE'] = df['CRASH_DATE'].apply(lambda x: pd.to_datetime(x))
```

### Tablas dinamicas

#### Opción 1

```python
df.groupby(['LIGHTING_CONDITION', 'REPORT_TYPE', 'CRASH_HOUR'])[['BEAT_OF_OCCURRENCE']].sum()
```

#### Opción 2

```python
df.groupby(['LIGHTING_CONDITION', 'REPORT_TYPE', 'CRASH_HOUR']).agg({'BEAT_OF_OCCURRENCE':'sum'})
```

#### Opción 3

```python
pd.pivot_table(df, index=['LIGHTING_CONDITION', 'REPORT_TYPE'])
```

### Aplicando filtros

```python
df.filter(['LIGHTING_CONDITION', 'REPORT_TYPE'])
```

### Tablas cruzadas

```python
pd.crosstab(df['LIGHTING_CONDITION'], df['CRASH_DAY_OF_WEEK'])
```
